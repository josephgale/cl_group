```python
count = 1994
while count <= 2021:
    with open(str(count)+'_environment.csv') as csv_file:
        for line in csv_file:
            print(str(count)+'_environment.csv has ' + str(line.count(',')+1) + ' entries.')
    count += 1
    
    
```
